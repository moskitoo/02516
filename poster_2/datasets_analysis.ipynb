```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def visualize_image(path):
    image = mpimg.imread(path)

    # Display the image
    plt.imshow(image)
    plt.axis('off')  # Turn off axis labels
    plt.show()

    print(image.shape)
```
